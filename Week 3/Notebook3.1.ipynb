## Random Walk

```python
import requests
from bs4 import BeautifulSoup
import datetime
import random
import re

random.seed(datetime.datetime.now().timestamp())

headers = {
    "User-Agent": "Mozilla/5.0"
}

def getLinks(articleUrl):
    url = f"https://en.wikipedia.org{articleUrl}"
    response = requests.get(url, headers=headers)
    bs = BeautifulSoup(response.text, "html.parser")
    return bs.find("div", {"id": "bodyContent"}).find_all(
        "a", href=re.compile("^(/wiki/)((?!:).)*$")
    )

links = getLinks("/wiki/Kevin_Bacon")

# Limit to 10 pages so it doesn't run forever
for i in range(10):
    if len(links) == 0:
        break

    newArticle = random.choice(links)["href"]
    print(newArticle)
    links = getLinks(newArticle)
```

## Recursively crawling an entire site

```python
import requests
from bs4 import BeautifulSoup
import re

pages = set()

headers = {
    "User-Agent": "Mozilla/5.0"
}

def getLinks(pageUrl, depth=0, max_depth=2):
    if depth > max_depth:
        return

    url = f"https://en.wikipedia.org{pageUrl}"
    response = requests.get(url, headers=headers)
    bs = BeautifulSoup(response.text, "html.parser")

    for link in bs.find_all("a", href=re.compile("^(/wiki/)")):
        if "href" in link.attrs:
            newPage = link.attrs["href"]

            if newPage not in pages:
                print(newPage)
                pages.add(newPage)
                getLinks(newPage, depth + 1, max_depth)

getLinks("")
```

## Collecting Data Across an Entire Site

```python
import requests
from bs4 import BeautifulSoup
import re

pages = set()

headers = {
    "User-Agent": "Mozilla/5.0"
}

def getLinks(pageUrl, depth=0, max_depth=2):
    if depth > max_depth:
        return

    url = f"https://en.wikipedia.org{pageUrl}"
    response = requests.get(url, headers=headers)
    bs = BeautifulSoup(response.text, "html.parser")

    try:
        print(bs.h1.get_text())

        bodyContent = bs.find("div", {"id": "bodyContent"}).find_all("p")
        if len(bodyContent):
            print(bodyContent[0].get_text(strip=True))

        edit_link = bs.find(id="ca-edit")
        if edit_link:
            print(edit_link.find("a")["href"])

    except AttributeError:
        print("This page is missing something! Continuing.")

    for link in bs.find_all("a", href=re.compile("^(/wiki/)")):
        newPage = link["href"]

        if newPage not in pages:
            print("-" * 20)
            print(newPage)
            pages.add(newPage)
            getLinks(newPage, depth + 1, max_depth)

getLinks("/wiki/General-purpose_programming_language")
```

## Crawling across the Internet

```python
import requests
from urllib.parse import urlparse, urljoin
from bs4 import BeautifulSoup
import random

headers = {"User-Agent": "Mozilla/5.0"}

def getSoup(url):
    response = requests.get(url, headers=headers, timeout=10)
    return BeautifulSoup(response.text, "html.parser")

def getInternalLinks(bs, url):
    netloc = urlparse(url).netloc
    internalLinks = set()

    for link in bs.find_all("a", href=True):
        full_url = urljoin(url, link["href"])
        if urlparse(full_url).netloc == netloc:
            internalLinks.add(full_url)

    return list(internalLinks)

def getExternalLinks(bs, url):
    netloc = urlparse(url).netloc
    externalLinks = set()

    for link in bs.find_all("a", href=True):
        full_url = urljoin(url, link["href"])
        if urlparse(full_url).netloc != netloc:
            externalLinks.add(full_url)

    return list(externalLinks)

def followExternalOnly(startingSite, steps=5):
    currentSite = startingSite

    for i in range(steps):
        bs = getSoup(currentSite)
        externalLinks = getExternalLinks(bs, currentSite)

        if externalLinks:
            currentSite = random.choice(externalLinks)
            print(f"Step {i+1}: {currentSite}")
        else:
            print("No external links found.")
            break

followExternalOnly("https://www.oreilly.com/", steps=5)
```

## Collect all External Links from a Site

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse

headers = {"User-Agent": "Mozilla/5.0"}

allExtLinks = []
allIntLinks = []

def getSoup(url):
    response = requests.get(url, headers=headers, timeout=10)
    return BeautifulSoup(response.text, "html.parser")

def getAllExternalLinks(url, max_pages=10):
    if len(allIntLinks) >= max_pages:
        return

    bs = getSoup(url)

    internalLinks = getInternalLinks(bs, url)
    externalLinks = getExternalLinks(bs, url)

    for link in externalLinks:
        if link not in allExtLinks:
            allExtLinks.append(link)
            print("External:", link)

    for link in internalLinks:
        if link not in allIntLinks:
            allIntLinks.append(link)
            print("Internal:", link)
            getAllExternalLinks(link, max_pages)

allIntLinks.append("https://www.oreilly.com/")
getAllExternalLinks("https://www.oreilly.com/", max_pages=10)
```
