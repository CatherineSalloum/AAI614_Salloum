```python
import pandas as pd

# URL of the webpage
url = "https://www.canada.ca/en/public-health/services/surveillance/respiratory-virus-detections-canada/2018-2019/respiratory-virus-detections-isolations-week-01-ending-january-5-2019.html"

# Read all tables
tables = pd.read_html(url)

# Select Table 2 (change the index if needed)
df = tables[1]

# Clean the DataFrame
df.columns = df.columns.astype(str).str.strip()
df = df.dropna(how="all").reset_index(drop=True)

# Remove summary rows such as Canada or Total (if present)
first_col = df.columns[0]
df = df[~df[first_col].astype(str).str.contains("Canada|Total", case=False, na=False)]

# Display the final DataFrame
print(df)

# Optional: save to CSV
df.to_csv("FluWatch_Table2.csv", index=False)

print("\nThe DataFrame has been created successfully!")
```
